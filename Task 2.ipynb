```python
con = PostgresConnection().getConnection()

# Load data from Excel file to pandas DataFrame
#df = pd.read_excel('data.xlsx')

# Rename column in DataFrame
df = Item_dim.rename(columns={'_desc': 'description'})

# Connect to database
#conn = psycopg2.connect(database='ecomdb', user='username', password='password', host='127.0.0.1', port='5432')

# Create cursor object
cur = con.cursor()

# Insert data into database
for index, row in df.iterrows():
    cur.execute('INSERT INTO ecomdb.Item_dim (item_key, item_name, description, unit_price, man_country, supplier, unit) VALUES (%s, %s, %s, %s, %s, %s, %s)', (row['item_key'], row['item_name'], row['description'], row['unit_price'], row['man_country'], row['supplier'], row['unit']))
    conn.commit()

# Close cursor and connection
cur.close()
conn.close()
```
